```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding, Bidirectional, Input, Concatenate
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from transformers import DistilBertTokenizer, TFDistilBertModel
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import pennylane as qml

# Load dataset
file_path = "/content/mustard-dataset.csv"
df = pd.read_csv(file_path)

# ----------------- VALUE TRACKING SECTION 1: ORIGINAL DATA -----------------
print("\n===== ORIGINAL DATA VALUES =====")
# Sample 3 random rows to track through the entire process
sample_indices = df.sample(n=3, random_state=42).index.tolist()
original_samples = df.loc[sample_indices].copy()
print("\nOriginal Sample Data (3 random rows):")
print(original_samples[['SENTENCE', 'SPEAKER', 'SHOW', 'SARCASM', 'SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']].to_string())
print("\nData types:")
print(df.dtypes[['SPEAKER', 'SHOW', 'SARCASM', 'SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']])
print("\nValue distributions:")
for col in ['SARCASM', 'SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT']:
    print(f"{col} - min: {df[col].min()}, max: {df[col].max()}, mean: {df[col].mean():.4f}")

# Drop rows with missing values
df = df.dropna(subset=['SARCASM', 'SENTENCE', 'SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT'])

# Encode categorical features
label_encoders = {}
for col in ['SPEAKER', 'SHOW', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Convert target variables
df['SARCASM'] = df['SARCASM'].astype(int)

# ----------------- VALUE TRACKING SECTION 2: AFTER ENCODING -----------------
print("\n===== AFTER CATEGORICAL ENCODING =====")
print("\nSample Data After Encoding:")
encoded_samples = df.loc[sample_indices].copy()
print(encoded_samples[['SPEAKER', 'SHOW', 'SARCASM', 'SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']].to_string())
print("\nEncoding mappings:")
for col in ['SPEAKER', 'SHOW', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']:
    if sample_indices[0] in df.index:
        original_value = original_samples.loc[sample_indices[0], col]
        encoded_value = encoded_samples.loc[sample_indices[0], col]
        print(f"{col}: '{original_value}' → {encoded_value}")

# Add feature engineering - key for improving sarcasm detection
df['SENTIMENT_DIFF'] = abs(df['SENTIMENT_IMPLICIT'] - df['SENTIMENT_EXPLICIT'])
df['EMOTION_MATCH'] = (df['EMOTION_IMPLICIT'] == df['EMOTION_EXPLICIT']).astype(int)
df['SENTIMENT_CONTRAST'] = ((df['SENTIMENT_IMPLICIT'] > 0.5) != (df['SENTIMENT_EXPLICIT'] > 0.5)).astype(int)

# ----------------- VALUE TRACKING SECTION 3: AFTER FEATURE ENGINEERING -----------------
print("\n===== AFTER FEATURE ENGINEERING =====")
engineered_samples = df.loc[sample_indices].copy()
print("\nSample Data After Feature Engineering:")
print(engineered_samples[['SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT', 'SENTIMENT_DIFF', 'EMOTION_MATCH', 'SENTIMENT_CONTRAST']].to_string())
print("\nFeature engineering example calculation:")
for idx in sample_indices[:1]:  # Just show for the first sample
    if idx in df.index:
        implicit = df.loc[idx, 'SENTIMENT_IMPLICIT']
        explicit = df.loc[idx, 'SENTIMENT_EXPLICIT']
        print(f"SENTIMENT_DIFF: |{implicit:.4f} - {explicit:.4f}| = {abs(implicit-explicit):.4f}")
        print(f"EMOTION_MATCH: {df.loc[idx, 'EMOTION_IMPLICIT']} == {df.loc[idx, 'EMOTION_EXPLICIT']} → {df.loc[idx, 'EMOTION_MATCH']}")
        print(f"SENTIMENT_CONTRAST: ({implicit:.4f} > 0.5) != ({explicit:.4f} > 0.5) → {df.loc[idx, 'SENTIMENT_CONTRAST']}")

# Load DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')

# Tokenize sentences - use entire sequence output, not just CLS token
max_len = 60
tokens = tokenizer(df['SENTENCE'].tolist(), padding='max_length', max_length=max_len, truncation=True, return_tensors='tf')
bert_output = bert_model(tokens['input_ids'], attention_mask=tokens['attention_mask'])[0]

# ----------------- VALUE TRACKING SECTION 4: BERT TEXT ENCODING -----------------
print("\n===== BERT TEXT ENCODING =====")
for idx in sample_indices[:1]:  # Just show for the first sample
    if idx in df.index:
        sample_idx = df.index.get_loc(idx)
        original_text = original_samples.loc[idx, 'SENTENCE']
        sample_tokens = tokens['input_ids'][sample_idx].numpy()
        sample_bert_output = bert_output[sample_idx].numpy()

        print(f"\nOriginal text: '{original_text}'")
        print(f"Token IDs (first 10): {sample_tokens[:10]}")
        print(f"Token decoded (first 5): {tokenizer.decode(sample_tokens[:5])}")
        print(f"BERT embedding shape: {sample_bert_output.shape}")
        print(f"BERT embedding first token (first 5 values): {sample_bert_output[0, :5]}")
        print(f"BERT embedding statistics: min={np.min(sample_bert_output):.4f}, max={np.max(sample_bert_output):.4f}, mean={np.mean(sample_bert_output):.4f}")

# Get speaker features - use one-hot encoding for better representation
speakers = pd.get_dummies(df['SPEAKER'], prefix='speaker')
shows = pd.get_dummies(df['SHOW'], prefix='show')

# ----------------- VALUE TRACKING SECTION 5: ONE-HOT ENCODING -----------------
print("\n===== ONE-HOT ENCODING =====")
for idx in sample_indices[:1]:  # Just show for the first sample
    if idx in df.index:
        sample_position = df.index.get_loc(idx)
        speaker_id = df.loc[idx, 'SPEAKER']
        show_id = df.loc[idx, 'SHOW']

        print(f"\nSpeaker ID: {speaker_id} → One-hot encoding:")
        speaker_one_hot = speakers.iloc[sample_position].values
        print(f"Speaker one-hot (first 10 positions): {speaker_one_hot[:10]}")
        print(f"Speaker one-hot: position {speaker_id} = 1, sum = {np.sum(speaker_one_hot)}")

        print(f"\nShow ID: {show_id} → One-hot encoding:")
        show_one_hot = shows.iloc[sample_position].values
        print(f"Show one-hot (first 10 positions): {show_one_hot[:10]}")
        print(f"Show one-hot: position {show_id} = 1, sum = {np.sum(show_one_hot)}")

# Select numerical features with engineered features
X_numerical = df[['SENTIMENT_IMPLICIT', 'SENTIMENT_EXPLICIT', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT',
                 'SENTIMENT_DIFF', 'EMOTION_MATCH', 'SENTIMENT_CONTRAST']]
scaler = StandardScaler()
X_numerical_scaled = scaler.fit_transform(X_numerical)

# ----------------- VALUE TRACKING SECTION 6: SCALING -----------------
print("\n===== STANDARDIZATION SCALING =====")
for idx in sample_indices[:1]:  # Just show for the first sample
    if idx in df.index:
        sample_position = df.index.get_loc(idx)

        print("\nNumerical features before scaling:")
        before_scaling = X_numerical.iloc[sample_position].values
        print(before_scaling)

        print("\nScaler parameters:")
        print(f"Mean values: {scaler.mean_}")
        print(f"Standard deviations: {scaler.scale_}")

        print("\nNumerical features after scaling:")
        after_scaling = X_numerical_scaled[sample_position]
        print(after_scaling)

        # Show scaling calculation for one feature
        feature_idx = 0
        feature_name = X_numerical.columns[feature_idx]
        orig_value = before_scaling[feature_idx]
        scaled_value = after_scaling[feature_idx]
        feature_mean = scaler.mean_[feature_idx]
        feature_std = scaler.scale_[feature_idx]

        print(f"\nExample scaling calculation for {feature_name}:")
        print(f"Original value: {orig_value}")
        print(f"Mean: {feature_mean}")
        print(f"Std: {feature_std}")
        print(f"Scaled value = (original - mean) / std = ({orig_value} - {feature_mean}) / {feature_std} = {scaled_value}")

# Add context features
context_features = np.hstack((X_numerical_scaled, speakers.values, shows.values))
y_sarcasm = df['SARCASM']

# Split data - stratify by sarcasm class to maintain distribution
X_bert_train, X_bert_test, X_context_train, X_context_test, y_train_sarc, y_test_sarc = train_test_split(
    bert_output.numpy(), context_features, y_sarcasm,
    test_size=0.2, random_state=42, stratify=y_sarcasm)

# Calculate class weights to handle imbalance
class_weights = compute_class_weight('balanced', classes=np.unique(y_train_sarc), y=y_train_sarc)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}

# ----------------- VALUE TRACKING SECTION 7: CLASS WEIGHTS -----------------
print("\n===== CLASS WEIGHTS FOR IMBALANCE =====")
print("\nClass distribution before SMOTE:")
class_counts = np.bincount(y_train_sarc.astype(int))
for cls, count in enumerate(class_counts):
    weight = class_weight_dict.get(cls, 0)
    print(f"Class {cls}: count={count}, proportion={count/len(y_train_sarc):.4f}, weight={weight:.4f}")

# Handle class imbalance with SMOTE for training data
print("Before SMOTE:", np.bincount(y_train_sarc.astype(int)))
X_bert_train_reshaped = X_bert_train.reshape(X_bert_train.shape[0], -1)  # Flatten BERT for SMOTE
combined_features = np.hstack((X_bert_train_reshaped, X_context_train))

smote = SMOTE(random_state=42)
combined_resampled, y_train_sarc_resampled = smote.fit_resample(combined_features, y_train_sarc)
print("After SMOTE:", np.bincount(y_train_sarc_resampled.astype(int)))

# ----------------- VALUE TRACKING SECTION 8: SMOTE RESAMPLING -----------------
print("\n===== SMOTE RESAMPLING =====")
print("\nClass distribution after SMOTE:")
class_counts_after = np.bincount(y_train_sarc_resampled.astype(int))
for cls, count in enumerate(class_counts_after):
    orig_count = class_counts[cls] if cls < len(class_counts) else 0
    print(f"Class {cls}: before={orig_count}, after={count}, new samples={count-orig_count}")

# Separate BERT and context features after SMOTE
bert_dim = X_bert_train_reshaped.shape[1]
X_bert_train_resampled = combined_resampled[:, :bert_dim].reshape(-1, max_len, 768)  # Reshape back to BERT format
X_context_train_resampled = combined_resampled[:, bert_dim:]

# Define quantum circuit
n_qubits = 6
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_circuit(inputs):
    # Encode inputs into rotation angles
    for i in range(n_qubits):
        qml.RY(inputs[i % len(inputs)], wires=i)

    # Entangle qubits (manually create a ring structure)
    for i in range(n_qubits):
        qml.CNOT(wires=[i, (i+1) % n_qubits])

    # Second layer of rotations
    for i in range(n_qubits):
        qml.RX(inputs[(i+1) % len(inputs)], wires=i)

    # More entanglement
    for i in range(0, n_qubits, 2):
        qml.CNOT(wires=[i, (i+1) % n_qubits])

    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# Apply quantum processing to a subset of features
def quantum_feature_extraction(inputs, feature_indices):
    # Select a subset of features to process quantum
    quantum_features = inputs[:, feature_indices]

    # Normalize features to suitable range for quantum rotations
    min_vals = np.min(quantum_features, axis=0)
    max_vals = np.max(quantum_features, axis=0)
    range_vals = max_vals - min_vals
    range_vals[range_vals == 0] = 1  # Avoid division by zero
    scaled_features = 2 * np.pi * (quantum_features - min_vals) / range_vals

    # Apply quantum circuit to each instance (batch in chunks to improve performance)
    batch_size = 100
    quantum_enhanced = []

    for i in range(0, len(scaled_features), batch_size):
        batch = scaled_features[i:i+batch_size]
        batch_output = np.array([quantum_circuit(instance) for instance in batch])
        quantum_enhanced.append(batch_output)

    return np.vstack(quantum_enhanced)

# ----------------- VALUE TRACKING SECTION 9: QUANTUM FEATURE EXTRACTION -----------------
print("\n===== QUANTUM FEATURE EXTRACTION =====")
# Extract emotional features for quantum processing
emotional_indices = [0, 1, 2, 3]  # sentiment and emotion indices

# Sample for demonstration (using just first few samples)
sample_quantum_input = X_context_train_resampled[:3]
quantum_features_sample = quantum_feature_extraction(sample_quantum_input, emotional_indices)

# Show the transformation steps
print("\nOriginal emotional features (first 3 samples):")
print(sample_quantum_input[:, emotional_indices])

# Calculate the normalization (2*π scaling) manually
quantum_input = sample_quantum_input[:, emotional_indices]
min_vals = np.min(quantum_input, axis=0)
max_vals = np.max(quantum_input, axis=0)
range_vals = max_vals - min_vals
range_vals[range_vals == 0] = 1
scaled_features = 2 * np.pi * (quantum_input - min_vals) / range_vals

print("\nScaled to 0-2π for quantum circuit (first 3 samples):")
print(scaled_features)

print("\nQuantum circuit outputs (first 3 samples):")
print(quantum_features_sample)

print("\nStatistics before quantum processing:")
print(f"Min: {np.min(quantum_input):.4f}, Max: {np.max(quantum_input):.4f}, Mean: {np.mean(quantum_input):.4f}")
print("\nStatistics after quantum processing:")
print(f"Min: {np.min(quantum_features_sample):.4f}, Max: {np.max(quantum_features_sample):.4f}, Mean: {np.mean(quantum_features_sample):.4f}")

# Apply quantum feature extraction (with progress output)
print("\nApplying quantum feature extraction to training data...")
quantum_features = quantum_feature_extraction(X_context_train_resampled, emotional_indices)
print("Quantum features shape:", quantum_features.shape)

# Add quantum features to context features
X_context_train_quantum = np.hstack((X_context_train_resampled, quantum_features))

print("Applying quantum feature extraction to test data...")
quantum_test_features = quantum_feature_extraction(X_context_test, emotional_indices)
X_context_test_quantum = np.hstack((X_context_test, quantum_test_features))

# Print information about the preprocessed datasets
print("\n--- FEATURE DIMENSIONS THROUGH PREPROCESSING ---")
print(f"Original numerical features: {X_numerical.shape[1]} dimensions")
print(f"After one-hot encoding speakers: +{speakers.shape[1]} dimensions")
print(f"After one-hot encoding shows: +{shows.shape[1]} dimensions")
print(f"After quantum feature extraction: +{quantum_features.shape[1]} dimensions")
print(f"Total context features: {X_context_train_quantum.shape[1]} dimensions")

# Print information about the preprocessed datasets
print("\n--- PREPROCESSED TRAINING DATA ---")
print(f"X_bert_train_resampled: shape={X_bert_train_resampled.shape}, dtype={X_bert_train_resampled.dtype}")
print(f"X_context_train_resampled: shape={X_context_train_resampled.shape}, dtype={X_context_train_resampled.dtype}")
print(f"X_context_train_quantum: shape={X_context_train_quantum.shape}, dtype={X_context_train_quantum.dtype}")
print(f"y_train_sarc_resampled: shape={y_train_sarc_resampled.shape}, dtype={y_train_sarc_resampled.dtype}")

print("\n--- PREPROCESSED TESTING DATA ---")
print(f"X_bert_test: shape={X_bert_test.shape}, dtype={X_bert_test.dtype}")
print(f"X_context_test: shape={X_context_test.shape}, dtype={X_context_test.dtype}")
print(f"X_context_test_quantum: shape={X_context_test_quantum.shape}, dtype={X_context_test_quantum.dtype}")
print(f"y_test_sarc: shape={y_test_sarc.shape}, dtype={y_test_sarc.dtype}")

# Print sample data (first few rows) for context features
print("\n--- SAMPLE DATA ---")
print("X_context_train_resampled (first 3 rows):")
print(X_context_train_resampled[:3])

print("\nX_context_train_quantum (first 3 rows - now with quantum features appended):")
print(X_context_train_quantum[:3])

print("\ny_train_sarc_resampled (first 10 elements):")
print(y_train_sarc_resampled[:10])

# Print class distribution in the preprocessed data
print("\n--- CLASS DISTRIBUTION ---")
print("Training data class distribution:")
print(np.bincount(y_train_sarc_resampled.astype(int)))
print("Testing data class distribution:")
print(np.bincount(y_test_sarc.astype(int)))

# ----------------- VALUE TRACKING SECTION 10: COMPLETE PIPELINE VISUALIZATION -----------------
print("\n===== COMPLETE PREPROCESSING PIPELINE =====")
# Choose one sample to track through the entire pipeline
if len(sample_indices) > 0 and sample_indices[0] in df.index:
    track_idx = sample_indices[0]
    sample_position = df.index.get_loc(track_idx)

    print(f"\nTracking sample through entire pipeline:")
    print(f"Original text: '{original_samples.loc[track_idx, 'SENTENCE']}'")
    print(f"Sarcasm label: {original_samples.loc[track_idx, 'SARCASM']}")

    print("\nPipeline steps:")
    print("1. Categorical encoding:")
    for col in ['SPEAKER', 'SHOW', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']:
        print(f"   {col}: '{original_samples.loc[track_idx, col]}' → {encoded_samples.loc[track_idx, col]}")

    print("\n2. Feature engineering:")
    for col in ['SENTIMENT_DIFF', 'EMOTION_MATCH', 'SENTIMENT_CONTRAST']:
        print(f"   {col}: {engineered_samples.loc[track_idx, col]}")

    print("\n3. Numerical feature scaling:")
    for i, col in enumerate(X_numerical.columns):
        orig_val = X_numerical.iloc[sample_position, i]
        scaled_val = X_numerical_scaled[sample_position, i]
        print(f"   {col}: {orig_val:.4f} → {scaled_val:.4f}")

    print("\n4. Quantum feature enhancement:")
    # Get the specific context features for this sample if it's in the training set
    # Note this is just an approximation since split/SMOTE changes indices
    print("   Original emotional features → Quantum enhanced features")
    print(f"   Values before: [limited sample representation]")
    print(f"   Values after: Emotional features processed through quantum circuit for 6 additional dimensions")

# Build an improved model architecture for sarcasm detection
# Text input branch for BERT output
text_input = Input(shape=(max_len, 768))
lstm_layer = Bidirectional(LSTM(192, return_sequences=True))(text_input)
lstm_layer2 = Bidirectional(LSTM(96))(lstm_layer)
lstm_output = Dropout(0.5)(lstm_layer2)

# Context input branch with quantum-enhanced features
context_input = Input(shape=(X_context_train_quantum.shape[1],))
context_dense1 = Dense(192, activation='relu')(context_input)
context_dense1 = Dropout(0.4)(context_dense1)
context_output = Dense(96, activation='relu')(context_dense1)

# Merge layers
merged = Concatenate()([lstm_output, context_output])
dense1 = Dense(256, activation='relu')(merged)
dense1 = Dropout(0.5)(dense1)
dense2 = Dense(128, activation='relu')(dense1)
dense2 = Dropout(0.4)(dense2)
dense3 = Dense(64, activation='relu')(dense2)
dense3 = Dropout(0.3)(dense3)
output = Dense(1, activation='sigmoid')(dense3)

# Create quantum-enhanced model
quantum_model = Model(inputs=[text_input, context_input], outputs=output)
quantum_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC()]
)

# Create baseline model (without quantum features) for comparison
# Text input branch for BERT output
baseline_text_input = Input(shape=(max_len, 768))
baseline_lstm_layer = Bidirectional(LSTM(192, return_sequences=True))(baseline_text_input)
baseline_lstm_layer2 = Bidirectional(LSTM(96))(baseline_lstm_layer)
baseline_lstm_output = Dropout(0.5)(baseline_lstm_layer2)

# Context input branch without quantum features
baseline_context_input = Input(shape=(X_context_train.shape[1],))
baseline_context_dense1 = Dense(192, activation='relu')(baseline_context_input)
baseline_context_dense1 = Dropout(0.4)(baseline_context_dense1)
baseline_context_output = Dense(96, activation='relu')(baseline_context_dense1)

# Merge layers
baseline_merged = Concatenate()([baseline_lstm_output, baseline_context_output])
baseline_dense1 = Dense(256, activation='relu')(baseline_merged)
baseline_dense1 = Dropout(0.5)(baseline_dense1)
baseline_dense2 = Dense(128, activation='relu')(baseline_dense1)
baseline_dense2 = Dropout(0.4)(baseline_dense2)
baseline_dense3 = Dense(64, activation='relu')(baseline_dense2)
baseline_dense3 = Dropout(0.3)(baseline_dense3)
baseline_output = Dense(1, activation='sigmoid')(baseline_dense3)

# Create and compile baseline sarcasm model
baseline_model = Model(inputs=[baseline_text_input, baseline_context_input], outputs=baseline_output)
baseline_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.AUC()]
)

# Training with callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy',
    patience=8,
    mode='max',
    restore_best_weights=True
)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.7,
    patience=4,
    min_lr=0.00001
)

# Train the baseline sarcasm model
print("Training baseline sarcasm model...")
baseline_history = baseline_model.fit(
    [X_bert_train_resampled, X_context_train_resampled],
    y_train_sarc_resampled,
    validation_data=([X_bert_test, X_context_test], y_test_sarc),
    epochs=40,
    batch_size=32,
    class_weight=class_weight_dict,
    callbacks=[early_stopping, reduce_lr]
)

# Train the quantum-enhanced model
print("Training quantum-enhanced model...")
quantum_history = quantum_model.fit(
    [X_bert_train_resampled, X_context_train_quantum],
    y_train_sarc_resampled,
    validation_data=([X_bert_test, X_context_test_quantum], y_test_sarc),
    epochs=40,
    batch_size=32,
    class_weight=class_weight_dict,
    callbacks=[early_stopping, reduce_lr]
)

# Evaluate models
baseline_results = baseline_model.evaluate([X_bert_test, X_context_test], y_test_sarc)
baseline_acc = baseline_results[1]
baseline_auc = baseline_results[2]

quantum_results = quantum_model.evaluate([X_bert_test, X_context_test_quantum], y_test_sarc)
quantum_acc = quantum_results[1]
quantum_auc = quantum_results[2]

print("\n--- FINAL RESULTS ---")
print(f"Baseline Model - Accuracy: {baseline_acc:.4f}, AUC: {baseline_auc:.4f}")
print(f"Quantum-Enhanced Model - Accuracy: {quantum_acc:.4f}, AUC: {quantum_auc:.4f}")

# Make predictions with quantum model
y_pred_prob = quantum_model.predict([X_bert_test, X_context_test_quantum])
y_pred = (y_pred_prob > 0.5).astype(int)

# Generate classification report
print("\nClassification Report (Quantum Model):")
print(classification_report(y_test_sarc, y_pred))

# Plot ROC curve
fpr, tpr, _ = roc_curve(y_test_sarc, y_pred_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, label=f'Quantum Model (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Sarcasm Detection')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Function to predict sarcasm on new text
def predict_sarcasm(text, speaker_id=0, show_id=0):
    # Tokenize and encode with BERT
    tokens = tokenizer(text, padding='max_length', max_length=max_len, truncation=True, return_tensors='tf')
    bert_features = bert_model(tokens['input_ids'], attention_mask=tokens['attention_mask'])[0].numpy()

    # Create speaker and show one-hot encodings
    speaker_encoding = np.zeros(speakers.shape[1])
    if speaker_id < speakers.shape[1]:
        speaker_encoding[speaker_id] = 1

    show_encoding = np.zeros(shows.shape[1])
    if show_id < shows.shape[1]:
        show_encoding[show_id] = 1

    # Create numerical features (neutral values)
    numerical = np.array([[0.5, 0.5, 0, 0, 0, 1, 0]])  # Add engineered features
    numerical = scaler.transform(numerical)

    # Combine context features
    context_features = np.hstack((numerical[0], speaker_encoding, show_encoding)).reshape(1, -1)

    # Add quantum features
    quantum_features = quantum_feature_extraction(context_features, emotional_indices)
    context_features_quantum = np.hstack((context_features, quantum_features))

    # Make prediction
    prediction = quantum_model.predict([bert_features, context_features_quantum])[0][0]
    print(f"Text: {text}")
    print(f"Sarcasm Probability: {prediction:.4f}")
    print(f"Prediction: {'Sarcastic' if prediction > 0.5 else 'Not Sarcastic'}")
    return prediction

# Test with examples
example_texts = [
    "I just love when my computer crashes right before a deadline.",
    "The weather is beautiful today.",
    "Oh great, another meeting that could have been an email.",
    "Thank you for your help, I really appreciate it."
]

for text in example_texts:
    predict_sarcasm(text)

# Compare quantum vs. baseline predictions
print("\n--- MODEL COMPARISON ---")
for text in example_texts:
    # Tokenize and process for both models
    tokens = tokenizer(text, padding='max_length', max_length=max_len, truncation=True, return_tensors='tf')
    bert_features = bert_model(tokens['input_ids'], attention_mask=tokens['attention_mask'])[0].numpy()

    # Create dummy features
    speaker_encoding = np.zeros(speakers.shape[1])
    show_encoding = np.zeros(shows.shape[1])
    numerical = np.array([[0.5, 0.5, 0, 0, 0, 1, 0]])
    numerical = scaler.transform(numerical)
    context_features = np.hstack((numerical[0], speaker_encoding, show_encoding)).reshape(1, -1)

    # Add quantum features
    quantum_features = quantum_feature_extraction(context_features, emotional_indices)
    context_features_quantum = np.hstack((context_features, quantum_features))

    # Get predictions
    baseline_pred = baseline_model.predict([bert_features, context_features])[0][0]
    quantum_pred = quantum_model.predict([bert_features, context_features_quantum])[0][0]

    print(f"\nText: {text}")
    print(f"Baseline Model - Sarcasm Probability: {baseline_pred:.4f} ({'Sarcastic' if baseline_pred > 0.5 else 'Not Sarcastic'})")
    print(f"Quantum Model - Sarcasm Probability: {quantum_pred:.4f} ({'Sarcastic' if quantum_pred > 0.5 else 'Not Sarcastic'})")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Input, Concatenate, Layer
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from transformers import DistilBertTokenizer, TFDistilBertModel
import pennylane as qml
import matplotlib.pyplot as plt

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load dataset
file_path = "/content/mustard-dataset.csv"
df = pd.read_csv(file_path)

# Drop rows with missing values
df = df.dropna(subset=['SENTENCE', 'SENTIMENT_IMPLICIT'])

# Encode categorical features
label_encoders = {}
for col in ['SPEAKER', 'SHOW', 'EMOTION_IMPLICIT', 'EMOTION_EXPLICIT']:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        label_encoders[col] = le

# Binarize sentiment for classification (1 = positive, 0 = negative)
# Assuming SENTIMENT_IMPLICIT values: negative sentiment < 0, positive sentiment > 0
df['SENTIMENT_BINARY'] = (df['SENTIMENT_IMPLICIT'] > 0).astype(int)

# Load DistilBERT tokenizer and model
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')

# Tokenize sentences
max_len = 50
sentences = df['SENTENCE'].tolist()
tokens = tokenizer(sentences, padding='max_length', max_length=max_len,
                   truncation=True, return_tensors='tf')
X_text = bert_model(tokens['input_ids'])[0][:, 0, :]  # Use [CLS] token embedding

# Select numerical features - using available emotional features
feature_cols = ['EMOTION_IMPLICIT', 'EMOTION_EXPLICIT'] if 'EMOTION_EXPLICIT' in df.columns else ['EMOTION_IMPLICIT']
if 'SENTIMENT_EXPLICIT' in df.columns:
    feature_cols.append('SENTIMENT_EXPLICIT')

X_numerical = df[feature_cols]
scaler = StandardScaler()
X_numerical = scaler.fit_transform(X_numerical)

# Combine features
X = np.hstack((X_text.numpy(), X_numerical))
y = df['SENTIMENT_BINARY']

# Dimensionality reduction for quantum processing
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
X_reduced = pca.fit_transform(X)

# Train-test split
X_train, X_test, X_reduced_train, X_reduced_test, y_train, y_test = train_test_split(
    X, X_reduced, y, test_size=0.2, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Reduced training data shape: {X_reduced_train.shape}")

# Define quantum circuit - without TensorFlow interface
n_qubits = 8
dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev)
def quantum_circuit(inputs, weights):
    # State preparation - encode inputs
    for i in range(n_qubits):
        qml.RY(inputs[i], wires=i)

    # Entanglement layers
    for layer in range(2):  # Two entanglement layers
        # Rotational layer with learnable weights
        for i in range(n_qubits):
            qml.RX(weights[layer][i][0], wires=i)
            qml.RY(weights[layer][i][1], wires=i)
            qml.RZ(weights[layer][i][2], wires=i)

        # Entanglement layer
        for i in range(n_qubits - 1):
            qml.CNOT(wires=[i, i + 1])
        qml.CNOT(wires=[n_qubits - 1, 0])  # Connect last qubit to first

    # Measure in Z basis
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

# Since we can't effectively integrate the quantum layer with Keras 3,
# let's use a simulated quantum approach (simplified version)
class SimulatedQuantumLayer(Layer):
    def __init__(self, output_dim=8, **kwargs):
        super(SimulatedQuantumLayer, self).__init__(**kwargs)
        self.output_dim = output_dim

    def build(self, input_shape):
        # Instead of quantum weights, we'll use classical weights that simulate
        # the behavior of a quantum circuit (nonlinear transformation)
        self.kernel = self.add_weight(
            name='kernel',
            shape=(input_shape[-1], self.output_dim),
            initializer='glorot_uniform',
            trainable=True
        )
        self.recurrent_weights = self.add_weight(
            name='recurrent_weights',
            shape=(self.output_dim, self.output_dim),
            initializer='orthogonal',
            trainable=True
        )
        self.bias = self.add_weight(
            name='bias',
            shape=(self.output_dim,),
            initializer='zeros',
            trainable=True
        )
        super(SimulatedQuantumLayer, self).build(input_shape)

    def call(self, inputs):
        # Apply nonlinear transformations that mimic quantum behavior
        # First linear transformation
        x = tf.matmul(inputs, self.kernel) + self.bias

        # Apply sine and cosine transformations (similar to quantum phases)
        x_sin = tf.sin(x)
        x_cos = tf.cos(x)

        # Mixing transformation (similar to entanglement)
        mixed = tf.matmul(x_sin, self.recurrent_weights) + x_cos

        # Apply tanh activation to bound output to [-1, 1] like quantum measurements
        return tf.tanh(mixed)

    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.output_dim)

# Build the hybrid model compatible with Keras 3
def build_hybrid_model(input_dim, reduced_dim):
    # Classical preprocessing branch
    classical_input = Input(shape=(input_dim,))
    classical_dense1 = Dense(64, activation='relu')(classical_input)
    classical_dropout1 = Dropout(0.3)(classical_dense1)
    classical_dense2 = Dense(32, activation='relu')(classical_dropout1)

    # Simulated Quantum branch
    quantum_input = Input(shape=(reduced_dim,))
    quantum_output = SimulatedQuantumLayer(output_dim=n_qubits)(quantum_input)
    quantum_dense = Dense(16, activation='relu')(quantum_output)

    # Concatenate classical and quantum branches
    concatenated = Concatenate()([classical_dense2, quantum_dense])

    # Fuzzy integration layer
    fuzzy_layer = Dense(16, activation='sigmoid')(concatenated)

    # Output layer
    output = Dense(1, activation='sigmoid')(fuzzy_layer)

    # Build model
    model = Model(inputs=[classical_input, quantum_input], outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    return model

# Build and train the model
model = build_hybrid_model(X_train.shape[1], X_reduced_train.shape[1])
print(model.summary())

# Train the model with early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy', patience=5, restore_best_weights=True)

# If batch size is too large, reduce it
batch_size = 16
if len(X_train) < 100:  # For very small datasets
    batch_size = 4

history = model.fit(
    [X_train, X_reduced_train], y_train,
    epochs=30,
    batch_size=batch_size,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1
)

# Evaluate the model
y_pred_proba = model.predict([X_test, X_reduced_test])
y_pred = (y_pred_proba > 0.5).astype(int)

print("\nModel Evaluation:")
print(f"Accuracy: {np.mean(y_pred.flatten() == y_test) * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.tight_layout()
plt.show()

# Function to predict sentiment of new sentences
def predict_sentiment(sentences, model, tokenizer, bert_model, pca, scaler, feature_cols, max_len):
    # Tokenize input sentences
    tokens = tokenizer(sentences, padding='max_length', max_length=max_len,
                     truncation=True, return_tensors='tf')
    text_features = bert_model(tokens['input_ids'])[0][:, 0, :]

    # Generate dummy numerical features (would need real features in production)
    dummy_numerical = np.zeros((len(sentences), len(feature_cols)))

    # Combine features
    X_new = np.hstack((text_features.numpy(), dummy_numerical))

    # Generate reduced features for quantum processing
    X_new_reduced = pca.transform(X_new)

    # Make prediction
    predictions = model.predict([X_new, X_new_reduced])

    results = []
    for i, sentence in enumerate(sentences):
        sentiment = "Positive" if predictions[i][0] > 0.5 else "Negative"
        confidence = predictions[i][0] if sentiment == "Positive" else 1 - predictions[i][0]
        results.append({
            "sentence": sentence,
            "sentiment": sentiment,
            "confidence": float(confidence) * 100
        })

    return results

# Example usage
test_sentences = [
    "I'm really happy with the results!",
    "This is absolutely terrible.",
    "The weather is quite nice today."
]

sentiment_results = predict_sentiment(test_sentences, model, tokenizer, bert_model, pca, scaler, feature_cols, max_len)
for result in sentiment_results:
    print(f"Sentence: \"{result['sentence']}\"")
    print(f"Sentiment: {result['sentiment']} (Confidence: {result['confidence']:.2f}%)")
    print()
```
